```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math
from dateutil.relativedelta import relativedelta
from alpha_vantage.fundamentaldata import FundamentalData
stock = 'CAT'
data = yf.download(stock, start='2022-04-01', end='2024-04-01', interval='1mo')
import warnings
warnings.filterwarnings('ignore')
'''
-Visual-based analysis of past two years of market data in one month ticks: include trading volume, earnings, and share price at minimum
-Three significant events that have occurred with the company in the past year
-The dominant shareholders of the company
-The goal / philosophy of the company
-A holistic analysis of whether the company is a good investment'''
```

<h1>Stock Price</h1>
<h4>In the last few months the price has skyrockted</h4>
<h5>In particular the last month has been insane</h5>

```python
fig, ax1 = plt.subplots(figsize=(8, 10)) 

# First plot: Stock price over time
i = 0
for index, row in data.iterrows():
    # Get the current date and calculate the new date by adding one month
    date = row["Date"]
    new_date = date + relativedelta(months=1)
    if i < len(data) - 1:
        next_start = data["Open"][i + 1]
    else:
        next_start = data["Close"][i]
    i += 1
    alpha_val = abs(row["Change"])/40
    alpha_val = min(alpha_val, 1)
    if row["Change"] > 0:
        ax1.plot([date, new_date], [row["Open"], next_start], color='green', alpha=alpha_val)
    else:
        ax1.plot([date, new_date], [row["Open"], next_start], color='red', alpha=alpha_val)
ax1.set_xlabel('Date')
ax1.set_ylim(150,350)
ax1.set_ylabel('Price')
ax1.set_title('Caterpillar Inc Stock Price Over Time')
ax1.legend(['Negative change', 'Positive change'])
```
